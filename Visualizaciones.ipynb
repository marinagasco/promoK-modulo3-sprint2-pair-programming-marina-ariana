```python
# primero nuestras librerias!

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
# lo primero que hacemos es cargar el dataframe que vamos a usar para esta lección
df = pd.read_csv("/Users/marinagasco/Desktop/VS_Bootcamp/Módulo_3/mod-3-pareja-2/promoK-modulo3-sprint2-pair-programming-marina-ariana/world_data_full_apply_nulos.csv", index_col=0) 
df.head(1)
```

```python
# 1. ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

sns.histplot(x = "birth_rate", data = df,  bins = 10, kde = True, color = "violet")

# usando 'plt.xlabel()' cambiamos el nombre del eje x
plt.xlabel("Años")

# usando el método 'plt.ylabel()' cambiamos el nombre del eje y
plt.ylabel("Natalidad");

# ponemos título a la gráfica
plt.title("Relación entre el tiempo transcurrido y la natalidad")

# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;
```

    Podemos obervar una tendencia descenciente de la natalidad a lo largo de los años. 
    Hay un repunte en el valor de años número 2, donde crece la natalidad seguramente ligado a un periodo de bonanza económica. 
    Hay una pico hacia abajo durante 2 décadas que quizás nos muestra periodo entre guerras etc. 

```python
# 2. ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

sns.scatterplot(x = 'infant_mortality', 
                y ='life_expectancy', 
                data = df)

# cambiamos los nombres de los ejes
plt.xlabel("Mortalidad infantil")
plt.ylabel("Esperanza de vida")

# ponemos título a la gráfica
plt.title("Relación entre la mortalidad infantil y la esperanza de vida")

# quitamos la linea de arriba y de la derecha
plt.gca().spines['right'].set_visible(False) # quitamos la línea de la derecha
plt.gca().spines["top"].set_visible(False) # quitamos la línea de arriba;
```

    Cuánta más espranza de vida, menps mortalidad infantil. 
    Y hay más cantidad de esta hipotesis que de: menos esperanza de vida, más mortalidad.

```python
# 3. ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

# empezamos preparando los datos para poder hacer el pieplot usando un groupby. En este caso tendremos que contar el número de clientes que tenemos por nivel educativo
df_países = df.groupby("continent")["country"].count().reset_index()
df_países
```

```python
# lo primero que vamos a hacer es definir una lista para definir los colores que queremos para cada una de las categorías. 
# Esta lista debe tener tantos elementos como categorías tenemos en la columna de estudio (en este caso 7)
colores = ["c", "cadetblue", "mediumseagreen", "seagreen", "teal", "turquoise", "lightseagreen"]


# creamos el gráfico basándonos en el DataFrame que hemos creado en el paso anterior
plt.pie("country", labels= "continent",
        data = df_países, 
        autopct=  '%1.1f%%', 
        colors = colores, 
        textprops={'fontsize': 8}, 
        startangle=90);
```

```python
df_africa = df[df["continent"] == "Africa"]
df_africa
```

```python
plt.pie(
    x=df_africa,  # Valores numéricos
    labels=df_africa["country"],  # Etiquetas (nombres de países)
    autopct='%1.1f%%', 
    colors = colores, 
    textprops={'fontsize': 8},
    startangle=90
)
```
